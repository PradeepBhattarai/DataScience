# Series

```python
import pandas as pd

x=[2,3,4,5,5,6]

var=pd.Series(x,index=['a','b','c','d','e','f'],dtype="float",name="python")
print(var)
print(type(var))
print(var[3])
```

```python
dic={"Name":['python','c','java','c++'],"por":[12,13,14,15],"rank":[1,4,3,2]}
var1=pd.Series(dic)
print(var1)

s=pd.Series(12,index=[1,2,3,4,5,6,7])
print(s)
print(type(s))

#no broadcasting error in pandas
s1=pd.Series(12,index=[1,2,3,4])
s2=pd.Series(8,index=[0,1,2])
print(s1+s2)
```

# DataFrames

```python
#2D data in pandas
l=[1,2,3,4,5,6]
var=pd.DataFrame(l)
print(var)
print(type(var))

d={"a":[1,2,3,4,5],"s":[6,7,8,9,10]}
var1=pd.DataFrame(d)
print(var1)
print(var1["a"][3],var1["s"][2])
print()

var1=pd.DataFrame(d,columns=["a"])
print(var1)
```

```python
list_1=[[1,2,3,4,5],[6,7,8,8,9]]
var2=pd.DataFrame(list_1)
print(type(list_1))
print(var2)
print()

sr={"s":pd.Series([1,2,3,4]),"r":pd.Series([4,5,6,7])}
var3=pd.DataFrame(sr)
print(var3)
print(type(var3))
```

# Arithmetic Operations in Pandas

```python
var=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})

var

var["C"]=var["A"]+var["B"]
var

var["Python"]=var["C"]<=8
var
```
